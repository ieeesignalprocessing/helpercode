## Plotting

```python
def fn1(t,f):
    return np.cos(2*np.pi*f*t)

def fn2(t,f):
    return np.sin(2*np.pi*f*t)
```

```python
t1 = np.linspace(0,5,500)
f1 = 5
t2 = np.linspace(0,5,10)
f2 = 10

plt.figure(1)
# plt.figure(2,(10,10))
plt.subplot(311)
plt.xlabel('t -->')
plt.ylabel('cos(t)')
plt.title('Cosine Plot')
plt.grid(True)
plt.plot(t1,fn1(t1,f1),'k',t2,fn1(t2,f2),'bo')
plt.text(2.1,0.8,r'$\pi, \sigma$')

plt.subplot(313)
plt.xlabel('t -->')
plt.ylabel('sin(t)')
plt.title('Sine Plot')
plt.plot(t1,fn2(t1,f2),'k',t2,fn2(t2,f2),'bo')
plt.legend(['Regular','Sampled'])
plt.show()
```

```python
t1 = np.linspace(0,5,500)
t2 = np.linspace(0,5,10)
f1 = 5
f2 = 2
# plt.figure(1)
# plt.figure(1,(20,10))
fig,ax = plt.subplots(2,1)
ax[0].plot(t1,fn1(t1,f1),'g',label = 'Cosine')
ax[1].plot(t1,fn2(t1,f2),'b',label = 'Sine')

ax[0].set_xlabel('t-->')
ax[0].set_ylabel('Y-->')
ax[0].set_title('Cosine function')

ax[1].set_xlabel('t-->')
ax[1].set_ylabel('Y-->')
ax[1].set_title('Sine function')

ax[1].grid()
ax[1].legend()

plt.show()
```
